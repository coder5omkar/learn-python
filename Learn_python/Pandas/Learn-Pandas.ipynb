```python
def get_city_match_count(city):
    mask=df['city'] == city
    return df[mask].shape[0]
```

```python
def get_city_match_count_and_duration(city):
    mask1=df['city'] == city
    mask2=df['date'] > '2017-01-01'
    return df[mask1 & mask2].shape[0]
```
