```python
## Write a code to check whether a given number is a palindrome.
```

```python
num = 101
original = num
rev = 0

while num > 0:
    digit = num % 10       # get digit
    rev = (rev * 10) + digit   # add digit to rev
    num = num//10  # divide number 
    
print(f"original number :{original} reverse number :{rev}")

if original == rev :
    print("palindrome number")
else :
    print("not palindrome number")
    
```

** Write a code to find the sum of numbers divisible by 4.The code must allow the user to accept a number and add it to the sum if it is divisible by 4. It should continue accepting numbers as long as the user wants to provide an input and should display the final sum.**

```python
total_sum = 0

while True:
    try :
        num = int(input("Enter a number: "))
        print(f"entered num {num}")

        if num%4 ==0:
            total_sum = total_sum +num
            print(f"adding number to sum :{total_sum}")
        else :
            print("number not divisible by 4")

        dicision = int(input("Enter 0 to continue 1 to exit: "))
        if dicision == 0:
            continue
        else :
            break
    except Exception:
        print("invalid number")
          
print(f"total sum is {total_sum}")
```
