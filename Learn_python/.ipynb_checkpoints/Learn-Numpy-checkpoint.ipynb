### Indexing with Boolean Array

```python
arr = np.random.randint(low=1,high=100,size=20).reshape(4,5)
```

```python
arr
```

```python
arr>50
```

```python
arr[(arr>50)&(arr%2 != 0)]
```

```python
arr[(arr>50)&(arr%2 != 0)] = 0
```

```python
arr
```
