```python
number = int(input("Enter number :"))
i = 1

while i<11:
    print(number, "*",i,"=", number*i)
    i+=1
```

## Nested Loops

```python
rows = int(input("Enter number :"))
```

```python
for i in range(1,rows+1):
    for j in range(0,i):
        print("*",end =" ")
    print("")
```

## break ,continue, past

```python
# break
for i in range(1,11):
    if i == 6:
        break
    print(i, end=" ")
```

```python
# continue
for i in range(1,11):
    if i == 6:
        continue
    print(i, end=" ")
```

```python
# pass
for i in range(1,10):
    pass
```
